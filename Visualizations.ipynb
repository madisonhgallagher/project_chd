```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/project_chd/clean_train.csv')

# Select relevant variables
vars = ['totChol', 'BMI', 'cigsPerDay', 'sysBP', 'diaBP', 'heartRate', 'glucose', 'TenYearCHD']
df_subset = df[vars]
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_subset.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

Based on the correlation heat map, it seems sysBP and diaBP are strongly correlated in the positive direction. As sysBP increases, so does diaBP. This makes sense as they both are variables that regard blood pressure so no matter what they are going to be somewhat correlated. It also seems like cigsPerDay and BMI are slightly negatively correlated which is interesting. As the number of cigarettes smoked per day increases, there's a slight tendency for BMI to decrease, and vice versa, though the relationship is not very strong.

```python
# Scatterplots
sns.pairplot(df_subset, kind='scatter', hue='TenYearCHD', plot_kws={'alpha': 0.5})
plt.suptitle('Pairplot of Predictors vs. TenYearCHD')
plt.show()
```

```python
# Boxplots
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(18, 10))
for i, var in enumerate(vars[:-1]):
    sns.boxplot(x='TenYearCHD', y=var, data=df, ax=axes[i//4, i%4])
    axes[i//4, i%4].set_title(f'Boxplot of {var} by TenYearCHD')
plt.tight_layout()
plt.show()
```

```python
# Violin Plots
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(18, 10))
for i, var in enumerate(vars[:-1]):
    sns.violinplot(x='TenYearCHD', y=var, data=df, ax=axes[i//4, i%4])
    axes[i//4, i%4].set_title(f'Violin Plot of {var} by TenYearCHD')
plt.tight_layout()
plt.show()
```
